3. Задание (в программе)

Напишите код на Python, реализующий построение графиков:
   
   1 окружности,

```python
x = []
y1 = []
y2 =[]
r = 5
for i in range(-100,101):
    x_ = i * r / 100
    x.append(x_)
    y1.append(np.sqrt(r**2 - x_ **2))
    y2.append(-np.sqrt(r**2 - x_ **2))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```

2.	эллипса,

```python
x = []
y1 = []
y2 =[]
a = 10
b = 4
for i in range(-100,101):
    x_ = i * a / 100
    x.append(x_)
    y1.append(np.sqrt(b**2 - (x_ * b / a)**2))
    y2.append(-np.sqrt(b**2 - (x_ * b / a)**2))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```

3.	гиперболы.

```python
x1 = []
x2 = []
y1 = []
y2 = []
a = 10
b = 2
for i in range(10,100):
    x_ = i*a/10
    x1.append(x_)
    x2.append(-x_)
    y1.append(np.sqrt(((x_/a)**2-1)*(b**2)))
    y2.append(-np.sqrt(((x_/a)**2-1)*(b**2)))
plt.plot(x1,y1)
plt.plot(x1,y2)
plt.plot(x2,y1)
plt.plot(x2,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

5. Задание (в программе)

1)	Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 3 * X + 5 * Y
Z1 = 3 * X + 5 * Y + 20
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z1)
plt.show()
```

2)	Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 3 * X ** 2 + 5 * Y
Z1 = np.sqrt(90 + 23 * X ** 2 - 0.5 * Y ** 2)
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z1)
plt.show()
```
