Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.

Суть ортогонального преобразования - это линейный перенос  (+ поворот). 

Если мы говорим про неискаженные поверхности, то любые фигуры при данных преобразованиях остаются в неизменном виде. Соответственно, и расстояния между любыми точками фигурами сохраняются.


Например: Отрезок с точками а(х1, у1) и с(х2, у2).
Расстояние между ними: ((х2-х1)**2 + (у2 - у1)**2)**0.5


Если сместили по оси х на к, по оси у на m, то:
a1(x1 + k, y1 +m), c(x2 + k, y2 + m)


Расстояние между ними: ((х2 + k - х1 - k)**2 + (у2 + m - у1 - m)**2)**0.5


Что абсолютно тождественно первому расстоянию.

1)	Напишите код, который будет переводить полярные координаты в декартовы.

```python
def dekart(x, y):
    r = np.sqrt(x**2 + y**2)
    alfa_ = np.arccos(x/r)
    return(r, alfa_)
```

```python
dekart(1, 2)
```

2)	Напишите код, который будет рисовать график окружности в полярных координатах.

```python
r = 5
alfa_ = 2
x = np.linspace(0, 2 * np.pi, 100)
y = (np.sqrt(r**2 - (x*np.cos(alfa_))**2))/np.sin(alfa_)
plt.polar(x, y)
plt.show()
```

1)	Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1


```python
from scipy.optimize import fsolve

x = np.linspace(-2, 3, 201)
#y = x**2 -1
#y1 = (np.exp(x) + x - 1)/x
plt.plot(x, x**2 -1)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot([-2,3], [0,0], color ='red', linewidth=2, linestyle ="--")
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2, 7)
plt.grid(True)
plt.show()
```

```python
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) - x*y + x -1)
x1, y1 = fsolve(equations, (-5, -5))
x2, y2 = fsolve(equations, (2, 3))
print(f'x1 = {x1:.3f}, y1 = {y1:.3f}')
print(f'x2 = {x2:.3f}, y2 = {y2:.3f}')
```
